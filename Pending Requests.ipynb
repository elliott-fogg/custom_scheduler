```python
request_groups = []
html_dir = "../request_generation/request_page_html"
with open(os.path.join(html_dir, "test_1.mhtml"), "r") as f:
    text = f.read()

soup = bs(text)
tbody_list = soup.findAll("tbody")
tbody = tbody_list[0]
divs = tbody.findAll("div", {"class": "row position-relative"})
print(len(divs))
for d in divs:
    request_time = d.findAll("span")[0].text.strip()
    total_requests = d.findAll("div", {"class": "total-requests"})[0].text.strip()
    pending_requests = d.findAll("div", {"class": "pending-requests"})[0].text.strip()
    failed_requests = d.findAll("div", {"class": "failed-requests"})[0].text.strip()
    completed_requests = d.findAll("div", {"class": "completed-requests"})[0].text.strip()
    request_group_name = d.findAll("a", {"class": "requestgroup-title"})[0].text.strip()
    proposal = d.findAll("a", {"class": "proposal-display-code"})[0].text.strip()

    request_groups.append({
        "time": request_time,
        "name": request_group_name,
        "total": total_requests,
        "pending": pending_requests,
        "failed": failed_requests,
        "completed": completed_requests,
        "proposal": proposal
    })
```

```python
request_groups = []
html_dir = "../request_generation/request_page_html"

file_list = os.listdir(html_dir)

for filename in file_list:
    with open(os.path.join(html_dir, filename), "r") as f:
        text = f.read()
    
    soup = bs(text)
    tbody_list = soup.findAll("tbody")
    tbody = tbody_list[0]
    divs = tbody.findAll("div", {"class": "row position-relative"})
    print(len(divs))
    for d in divs:
        request_time = d.findAll("span")[0].text.strip()
        total_requests = d.findAll("div", {"class": "total-requests"})[0].text.strip()
        pending_requests = d.findAll("div", {"class": "pending-requests"})[0].text.strip()
        failed_requests = d.findAll("div", {"class": "failed-requests"})[0].text.strip()
        completed_requests = d.findAll("div", {"class": "completed-requests"})[0].text.strip()
        request_group_name = d.findAll("a", {"class": "requestgroup-title"})[0].text.strip()
        proposal = d.findAll("a", {"class": "proposal-display-code"})[0].text.strip()

        request_groups.append({
            "time": request_time,
            "name": request_group_name,
            "total": total_requests,
            "pending": pending_requests,
            "failed": failed_requests,
            "completed": completed_requests,
            "proposal": proposal
        })
```

```python
df = pd.DataFrame(request_groups)
df
```
