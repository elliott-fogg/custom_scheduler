## Load Request Data

```python
request_dfs = []
dirpath = "D:\OBSERVATION_DATA/requestgroup_raw"
for filename in os.listdir(dirpath):
    if filename.startswith("requestgroups_request_"):
        filepath = os.path.join(dirpath, filename)
        data = json.load(open(filepath, "r"))["requestgroups_request"]
        df = pd.DataFrame.from_dict(data)
        request_dfs.append(df[["id", "state", "created", "request_group_id", "optimization_type"]])
request_data = pd.concat(request_dfs)
request_data
```

## Load RequestGroup Data

```python
group_dfs = []
dirpath = "D:\OBSERVATION_DATA/requestgroup_raw"
for filename in os.listdir(dirpath):
    if filename.startswith("requestgroups_requestgroup_"):
        filepath = os.path.join(dirpath, filename)
        data = json.load(open(filepath, "r", encoding="utf8"))["requestgroups_requestgroup"]
        df = pd.DataFrame.from_dict(data)
        group_dfs.append(df[["id", "observation_type", "ipp_value", "proposal_id"]])
group_data = pd.concat(group_dfs)
group_data.rename(columns={"id": "request_group_id"}, inplace=True)
group_data
```

## Load Window Data

```python
window_dfs = []
dirpath = "D:\OBSERVATION_DATA/requestgroup_raw"
for filename in os.listdir(dirpath):
    if filename.startswith("requestgroups_window_"):
        filepath = os.path.join(dirpath, filename)
        data = json.load(open(filepath, "r", encoding="utf8"))["requestgroups_window"]
        df = pd.DataFrame.from_dict(data)
        window_dfs.append(df[["start", "end", "request_id"]])
window_data = pd.concat(window_dfs)
window_data
```

```python
request_data["creation_date"] = request_data["created"].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%dT%H:%M:%S.%fZ").date())

def find_semester_position(dt_date):
    semester_starts = [
        dt.date(2020, 2, 1),
        dt.date(2020, 8, 1),
        dt.date(2021, 2, 1),
        dt.date(2021, 8, 1),
        dt.date(2022, 2, 1),
        dt.date(2022, 8, 1)
    ]
    semester_starts.sort(reverse=True)
    for s in semester_starts:
        if dt_date > s:
            break
    return (dt_date - s).days
    
request_data["semester_position"] = request_data["creation_date"].apply(find_semester_position)

request_data["weekday"] = request_data["creation_date"].apply(lambda x: x.weekday())
request_data["week"] = request_data["creation_date"].apply(lambda x: x.isocalendar()[1])
```
