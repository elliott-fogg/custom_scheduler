```python
from generate_requests_v2 import *
import os
import time
import random
```

### Define Seed Generation Function

```python
def init_generate(seed):
    random.seed(seed)
    rg = RequestGeneratorV2(0, 15000, 5, 300, 1)
    rg.generate_input_params()
    rg.generate_single_request_injection(0, 5000, 60, 3000)
    return rg.output_to_json()
```

### Generate JSON files from seeds

```python
dirname = os.path.join("test_results", "seed_checks")
os.makedirs(dirname, exist_ok=True)
for i in range(5):
    filepath = os.path.join("test_results", "seed_checks", f"seed_check_{i}.json")
    output = init_generate(i)
    json.dump(output, open(filepath, "w"))
```

### Testing JSON generation

```python
for i in range(5):
    filepath = os.path.join("test_results", "seed_checks", f"seed_check_{i}.json")
    output1 = json.load(open(filepath, "r"))
    str1 = json.dumps(output1)
    output2 = init_generate(i)
    str2 = json.dumps(output2)
    print(f"Seed: {i}, Valid: {str1 == str2}")
```

### Testing JSON generation backwards

```python
for i in range(4, -1, -1):
    filepath = os.path.join("test_results", "seed_checks", f"seed_check_{i}.json")
    outputb1 = json.load(open(filepath, "r"))
    strb1 = json.dumps(outputb1)
    outputb2 = init_generate(i)
    strb2 = json.dumps(outputb2)
    print(f"Seed: {i}, Valid: {str1 == str2}")
```
