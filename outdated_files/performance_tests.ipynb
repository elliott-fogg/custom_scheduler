### Number of Networks
Experiments to determine whether the number of networks (collections of telescopes operating independently of one another) has an impact on the performance of the schedule. i.e. With the same number of telescopes, by splitting the network up into sections that don't interact with each other, is the optimiser able to interpret that for an improvement in performance that compares with just having separate schedulers running?

Hypothesis: It does not, having multiple "networks" running in the same algorithm still slows everything down, and they would benefit from splitting them up.

### Scaling Numbers of Networks
Experiments to see determine whether a schedule that has several independent networks in it performs simply as a multiple of the performance of a single instance of the network.
i.e. If we have a multi-network system of 3 independent networks, each composed of 4 telescopes, and with 500 requests per network, does it simply take (roughly) 3 times as long as a single network of 4 telescopes with 500 requests? Do we see the same scaling as we would expect just from increasing the number of requests and telescopes? Or does it lie somewhere in between?

```python
pt6 = tt.PerformanceTest6("test_results/performance",
                          [500, 1000],
                          [4, 10],
                          [5],
                          3)
```
